```python
import numpy as np
import pandas as pd

def generate_note_dataframe(songString):
    # Total rows needed
    total_rows = 42

    # Extract "note" and "next note" with proper wrapping
    notes = []
    next_notes = []

    for i in range(total_rows):
        # Calculate indices for the note and next note
        start_index = i % len(songString)
        end_index = (i + 4) % len(songString)
        next_note_index = (i + 4) % len(songString)

        # Check if the note wraps around the string end
        if end_index > start_index:
            note = songString[start_index:end_index]
        else:
            note = songString[start_index:] + songString[:end_index]

        # Next note is simply the character at the next_note_index
        next_note = songString[next_note_index]

        # Append to lists
        notes.append(note)
        next_notes.append(next_note)

    # Create DataFrame
    df = pd.DataFrame({
        "note": notes,
        "next note": next_notes
    })

    return df


```
